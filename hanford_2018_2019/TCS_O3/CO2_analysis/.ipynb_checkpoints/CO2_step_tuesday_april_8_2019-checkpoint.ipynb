```python
import numpy as np
import scipy.signal as sig
import nds2
import sys
import matplotlib.pyplot as plt
sys.path.append('../cust_python/')
from fetch_and_saveh5 import *
```

```python
chans = np.loadtxt('../tcs_chan_list.txt',dtype='string')
```

```python
chans_update = [chans[1], chans[3], chans[12],chans[13]]
print(chans_update)
```

```python
#fetch_and_saveh5('CO2_step_response',chans_update, 1238858067, 1238870468)
```

```python
fname = 'CO2_step_response'
conn = nds2.connection('h1nds1',8088)
t_start = 1238858067
t_end = 1238870468
data_h5 = conn.fetch(t_start, t_end, chans_update)
h5f = h5py.File('{}.h5'.format(fname), 'w')
for i in range(0,len(data_h5)):
    h5f.create_dataset(data_h5[i].name,data=data_h5[i].data)
    h5f[data_h5[i].name].attrs['sample_rate']=data_h5[i].sample_rate
h5f.close()
```

```python
h5f1 = h5py.File('{}.h5'.format(fname),'r')
names = h5f1.keys()
sample_rates = []
sample_rates = [h5f1[names[i]].attrs['sample_rate'] for i in range(0,len(names))]
CO2X = h5f1[names[0]]
HWSX = h5f1[names[1]]
CO2Y = h5f1[names[2]]            
HWSY = h5f1[names[3]]
```

```python
h5f1.keys()
```

```python
#time series
f_s = 16
b = len(CO2X)
t  = np.linspace(0, float(b)/float(f_s), b)
print(f_s)
```

```python
plt.plot(t,CO2X)
plt.plot(t,CO2Y)
```

```python
#Low pass for data conditioning
b, a = sig.butter(2, 0.000025)
b_1,a_1 = sig.butter(2,0.00007)
```

```python
fig = plt.figure(figsize=(15,10))
plt.subplot(211)
plt.plot(t,HWSX, label='raw HWSX')
plt.plot(t, sig.filtfilt(b,a,HWSX), label='LP HWSX')
plt.subplot(212)
plt.plot(t,HWSY, label='raw HWSY')
plt.plot(t, sig.filtfilt(b_1,a_1,HWSY), label='LP HWSY')
```
