```python
def pzt_fit(time):
  return p[0] * time**2 + p[1] * time + p[2]
```

```python
test_t = np.arange(-70,140,.1)
y = [pzt_fit(i) for i in test_t]
plt.plot(test_t,y)
```

```python
#Peak identification (characterizing the OMC)
FSR = 261710400.0
TMS = 57982266.0
fines = 375.5
```

```python
def P_circ(nu, fines, fsr, P_max=1): 
    P = P_max/( (1 + (2.0*fines/np.pi)**2 * np.sin((np.pi*nu)/fsr)**2) )
    return P
```

```python
P = [P_circ(i, fines, FSR, P_max=10.5) for i in y]
```

```python
plt.semilogy(y,P)
```

```python
peaks = sig.find_peaks(P, height= 8)
print(peaks)
(FSR-(y[peaks[0][2]] - y[peaks[0][1]]))/FSR
```

```python
o_n = 2
o_n2 = 1
y_1 = [pzt_fit(i) for i in t[a[0][0]:a[0][1]]]
mod_freq = 45e6

P_zero = [P_circ(k, fines, FSR, P_max=10.5) for k in y_1]
P_zerp1 = [P_circ(k, fines, FSR+TMS*o_n, P_max=10.5) for k in y_1]
P_pos = [P_circ(k, fines, FSR+mod_freq, P_max=10.5) for k in y_1]
P_pos1 = [P_circ(k, fines, FSR+mod_freq+TMS*o_n, P_max=10.5) for k in y_1]
P_neg = [P_circ(k, fines, FSR-mod_freq, P_max=10.5) for k in y_1]
P_neg1 = [P_circ(k, fines, FSR-mod_freq+TMS*o_n, P_max=10.5) for k in y_1]
P_pos2= [P_circ(k, fines, FSR+TMS*o_n2, P_max=10.5) for k in y_1]
P_zero2= [P_circ(k, fines, FSR+mod_freq+TMS*o_n2, P_max=10.5) for k in y_1]
P_neg2= [P_circ(k, fines, FSR-mod_freq+TMS*o_n2, P_max=10.5) for k in y_1]
```
