```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create dictionaries for "hot" and "cold" ifo states
COLD_STATE_POWER_DICT = {
        'TCS_ITMX_CO2_PWR' : 1.06,
        'TCS_ITMY_CO2_PWR' : 1.78
}

HOT_STATE_POWER_DICT = {
        'TCS_ITMX_CO2_PWR' : 0.633333,
        'TCS_ITMY_CO2_PWR' : 0.0
}

HOT_STATE_CO2X_POWER = HOT_STATE_POWER_DICT['TCS_ITMX_CO2_PWR']
HOT_STATE_CO2Y_POWER = HOT_STATE_POWER_DICT['TCS_ITMY_CO2_PWR']

COLD_STATE_CO2X_POWER = COLD_STATE_POWER_DICT['TCS_ITMX_CO2_PWR']
COLD_STATE_CO2Y_POWER = COLD_STATE_POWER_DICT['TCS_ITMY_CO2_PWR']

PSL_TARGET = 50.0
PSL_START = 2.0
```

```python
## IMPORTANT POINTS ##

# Dan's sideband balancing CO2 tuning (11/07/2018)
Dan_CO2X = [20, 0.9]
Dan_CO2Y = [20, 0.95]

# Suggested from Hartmann (based off of CO2 calibration and measurement of self-heating)
# TBD
```

```python
# Calculate slope for line that the CO2 power will go down (under "get_tcs_power" in guardian)
m_x = (HOT_STATE_CO2X_POWER - COLD_STATE_CO2X_POWER)/(PSL_TARGET - PSL_START)
m_y = (HOT_STATE_CO2Y_POWER - COLD_STATE_CO2Y_POWER)/(PSL_TARGET - PSL_START)
pow_step = .1
PSL_POW = np.arange(PSL_START,PSL_TARGET + pow_step,pow_step)
```

```python
# CO2X_POW vs. PSL_POW
CO2X_POW = [m_x*(i-2)  + COLD_STATE_CO2X_POWER for i in PSL_POW]
plt.plot(PSL_POW, CO2X_POW, label='curve in guardian')
plt.plot(Dan_CO2X[0], Dan_CO2X[1], '*', label='good sideband balance (12/07/2018)')
plt.xlabel('PSL POWER (W)')
plt.ylabel('CO2X POWER (W)')
plt.legend()
plt.title('CO2X reduction function')
plt.show()
print('@ PSL POWER {} W , CO2X POWER is {} W'.format(str(PSL_POW[len(PSL_POW)-1]), str(CO2X_POW[len(PSL_POW)-1])))
```

```python
# CO2Y_POW vs. PSL_POW
CO2Y_POW = [m_y*(i-2.0)  + COLD_STATE_CO2Y_POWER for i in PSL_POW]
plt.plot(PSL_POW, CO2Y_POW, label='curve in guardian')
plt.plot(Dan_CO2Y[0], Dan_CO2Y[1], '*', label='good sideband balance (12/07/2018)')
plt.xlabel('PSL POWER (W)')
plt.ylabel('CO2Y POWER (W)')
plt.title('CO2Y reduction function')
plt.legend()
plt.show()
print('@ PSL POWER {} W , CO2Y POWER is {} W'.format(str(PSL_POW[len(PSL_POW)-1]), str(CO2Y_POW[len(PSL_POW)-1])))
```
