# SRCL to DARM Feedforward Fitting
1) Take DARM ERR/SRCL OUT measurement

2) Take DARM ERR/SRCL FF measurement

3) Divide them

4) Fit them

5) Tune the fit

6) Put into SRCL FF

7) Win megaparsecs

```python
# Read in xmls
if not iterative_method :
    SRCL2DARMxml ='/opt/rtcds/userapps/release/lsc/h1/scripts/feedforward/SRCL_to_DARM_excitation.xml'
else: 
    SRCL2DARMxml = '/opt/rtcds/userapps/release/lsc/h1/scripts/feedforward/SRCL_to_DARM_excitation_iterative.xml'

SRCLFF2DARMxml ='/opt/rtcds/userapps/release/lsc/h1/scripts/feedforward/SRCLFF_to_DARM_excitation.xml'

SRCL2DARMdict = dtt2hdf.read_diaggui(SRCL2DARMxml)
SRCLFF2DARMdict = dtt2hdf.read_diaggui(SRCLFF2DARMxml)
SRCLFF2DARMdict['results']['CSD'].keys()
```

```python
# In Lee's dtt2hdf extraction code, PSDs are actually ASDs
coh_cutoff = 0.90
ff_low_cutoff = 20 # Hz
ff_high_cutoff = 200.0 # Hz

SRCLout = 'H1:LSC-SRCL_OUT_DQ'
SRCLFF = 'H1:LSC-SRCLFF1_IN2'
DARMerr = 'H1:LSC-DARM_IN1_DQ'

index = SRCL2DARMdict['results']['COH'][SRCLout]['channelB_inv'][DARMerr]
S2D_coh = SRCL2DARMdict['results']['COH'][SRCLout]['coherence'][index]
S2D_CSD = SRCL2DARMdict['results']['CSD'][SRCLout]['CSD'][index]
S2D_PSD = SRCL2DARMdict['results']['PSD'][SRCLout]['PSD'][0]**2
S2D_TF = S2D_CSD/S2D_PSD

FFindex = SRCLFF2DARMdict['results']['COH'][SRCLFF]['channelB_inv'][DARMerr]
SFF2D_coh = SRCLFF2DARMdict['results']['COH'][SRCLFF]['coherence'][FFindex]
SFF2D_CSD = SRCLFF2DARMdict['results']['CSD'][SRCLFF]['CSD'][FFindex]
SFF2D_PSD = SRCLFF2DARMdict['results']['PSD'][SRCLFF]['PSD'][0]**2
SFF2D_TF = SFF2D_CSD/SFF2D_PSD

ff = SRCL2DARMdict['results']['CSD']['H1:LSC-SRCL_OUT_DQ']['FHz']

# Since we have the same frequency vectors, find the freq points where coherence is > than the cutoff for both
goodCoh = np.intersect1d(np.argwhere(S2D_coh > coh_cutoff)[:,0], np.argwhere(SFF2D_coh > coh_cutoff)[:,0])
goodCoh = np.intersect1d(goodCoh, np.argwhere(ff > ff_low_cutoff))
goodCoh = np.intersect1d(goodCoh, np.argwhere(ff_high_cutoff > ff))

S2D_goodCoh = S2D_coh[goodCoh]
S2D_goodCSD = S2D_CSD[goodCoh]
S2D_goodPSD = S2D_PSD[goodCoh]
S2D_goodTF = S2D_goodCSD/S2D_goodPSD

SFF2D_goodCoh = SFF2D_coh[goodCoh]
SFF2D_goodCSD = SFF2D_CSD[goodCoh]
SFF2D_goodPSD = SFF2D_PSD[goodCoh]
SFF2D_goodTF = SFF2D_goodCSD/SFF2D_goodPSD


goodff = ff[goodCoh]

print 'Number of points with coherence above {}: {}'.format(coh_cutoff, len(goodCoh))
```

### Iterative Method
So we've got $-\dfrac{\frac{DARMIN}{MICH}}{\frac{DARMIN}{MICHFF}}$ fit from above.  Now we just read in the current FF settings $\alpha$ from the FM to compute to new additional FF $\beta$

```python
def tconvert(t):
    return int(subprocess.check_output(['tconvert', t]))
```

```python
gpstime = tconvert('now-3mins')
FMs = np.array(['LSC-SRCLFF1'])
FMTFs, FMdict, fotonDict, args = filterModuleStateExtractor.getFMState(['LSC-SRCLFF1'], gpstime)
```

```python
FMsActivated = np.array([])
for key in FMdict[FMs[0]]['ONOFF'].keys():
    if 'FM' in key and FMdict[FMs[0]]['ONOFF'][key] == '1':
        FMsActivated = np.append(FMsActivated, key)
print FMsActivated
```

```python
fs = fotonDict[FMs[0]]['fs']
wwNorm = 2 * np.pi * goodff / fs
wwNormfull = 2 * np.pi * ff / fs

alphaTF = np.array([])
alphaTFfull = np.array([])
for activatedFM in FMsActivated:
    key = int(activatedFM[-1])-1
    sos = fotonDict[FMs[0]][key]['sosCoeffs']
    ww, TF = sig.sosfreqz(sos, worN=wwNorm)
    ww, TFfull = sig.sosfreqz(sos, worN=wwNormfull)
    
    if 'Vio' not in fotonDict[FMs[0]][key]['name']:
        if len(alphaTF) == 0:
            alphaTF = np.copy(TF)
            alphaTFfull = np.copy(TFfull)
        else:
            alphaTF *= TF
            alphaTFfull *= TFfull
```
